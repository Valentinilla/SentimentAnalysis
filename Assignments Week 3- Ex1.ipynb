```python
def awesome_count(review_dict):
    if 'awesome' in review_dict:
        count = review_dict['awesome']
    else:
        count = 0
    return count
```

```python
awesome_count(products['word_count'][1])
```

```python
products['awesome'] = products['word_count'].apply(awesome_count)
```

```python
def word_occurrences(review):
    words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
    d = dict()
    for x in words:
        if x in review:
            d[x] = review[x]
        else:
            d[x] = 0
    return d
```

```python
print(products['word_count'].apply(word_occurrences))
```

```python
test = products['word_count'].apply(word_occurrences)
```

```python
for voc in test.dict_keys()[1]:
    products[voc] = test.apply(lambda x: x[voc])
```
